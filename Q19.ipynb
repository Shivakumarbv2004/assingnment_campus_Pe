```python
# QUESTION NO 19:


# TEXT ANALYSIS PROGRAM


# Step 1: Take input
text = input("Enter text: ")

# Step 2: Count words
words = text.split()
word_count = len(words)

# Step 3: Count vowels and consonants
vowels = "aeiouAEIOU"
vowel_count = sum(1 for ch in text if ch in vowels)
consonant_count = sum(1 for ch in text if ch.isalpha() and ch not in vowels)

# Step 4: Reverse text
reversed_text = text[::-1]

# Step 5: Check palindrome
is_palindrome = text.lower() == reversed_text.lower()

# Step 6: Remove vowels
without_vowels = "".join(ch for ch in text if ch not in vowels)

# Step 7: Display results
print("\n----- TEXT ANALYSIS -----")
print("Words        :", word_count)
print("Vowels       :", vowel_count)
print("Consonants   :", consonant_count)
print("Reversed     :", reversed_text)
print("Palindrome   :", is_palindrome)
print("Without Vowels:", without_vowels)
```
