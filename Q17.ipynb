```python
# QUESTION NO 17:


# PALINDROME CHECKER PROGRAM


# Step 1: Take input
text = input("Enter word or number: ")

# Step 2: Reverse text
reversed_text = text[::-1]

# Step 3: Compare ignoring case
if text.lower() == reversed_text.lower():
    result = "PALINDROME"
else:
    result = "NOT PALINDROME"

# Step 4: Display result
print("Original :", text)
print("Reversed :", reversed_text)
print("Result   :", result)
```
