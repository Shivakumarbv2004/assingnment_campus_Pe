```python
# QUESTION NO 8:  


# LEAP YEAR CHECKER PROGRAM

# Step 1: Take year input
year = int(input("Enter a year: "))

# Step 2: Apply leap year condition
if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    result = "Leap Year"
else:
    result = "Not a Leap Year"

# Step 3: Display result
print("\nYear :", year)
print("Result :", result)
```
