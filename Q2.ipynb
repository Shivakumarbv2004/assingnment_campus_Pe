```python
# QUESTION NO 2:


# SIMPLE CALCULATOR PROGRAM

# Step 1: Take two numbers as input from the user
# float() is used so the program can handle decimal values also
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Step 2: Perform arithmetic operations

addition = num1 + num2              # Adds two numbers
subtraction = num1 - num2           # Subtracts second number from first
multiplication = num1 * num2        # Multiplies two numbers

# Division and modulus require checking division by zero
if num2 != 0:
    division = num1 / num2          # Divides first number by second
    modulus = num1 % num2           # Gives remainder
else:
    division = "Undefined (cannot divide by zero)"
    modulus = "Undefined (cannot divide by zero)"

exponentiation = num1 ** num2       # Raises num1 to the power of num2

# Step 3: Display the results neatly
print("\n----- CALCULATOR RESULTS -----")
print("Addition        :", addition)
print("Subtraction     :", subtraction)
print("Multiplication  :", multiplication)
print("Division        :", division)
print("Modulus         :", modulus)
print("Exponentiation  :", exponentiation)
```
