```python
# QUESTION NO 1:

# STUDENT BIO CARD PROGRAM


# Step 1: Store student details in variables
name = input("enter the name")
age = int(input("enter the age"))
course = input("course you are intrested")
college = input("enter the college name")
email = input("enter your mail")

# Step 2: Print the top border of the card
print("╔════════════════════════════════╗")

# Step 3: Print the title centered inside the box
# .center(32) is used to align text in the middle
print("║" + " STUDENT BIO CARD ".center(32) + "║")

# Step 4: Print a separator line
print("╠════════════════════════════════╣")

# Step 5: Print each detail inside the box
# .ljust(32) is used to align text to the left
print("║" + f" Name    : {name}".ljust(32) + "║")
print("║" + f" Age     : {age} years".ljust(32) + "║")
print("║" + f" Course  : {course}".ljust(32) + "║")
print("║" + f" College : {college}".ljust(32) + "║")
print("║" + f" Email   : {email}".ljust(32) + "║")

# Step 6: Print the bottom border of the card
print("╚════════════════════════════════╝")

```
