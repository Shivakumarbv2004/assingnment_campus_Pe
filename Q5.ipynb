```python
#QUESTION NO 5:


# BILL SPLITTER PROGRAM


# Step 1: Take inputs from user
total_bill = float(input("Enter total bill amount: ₹"))
number_of_people = int(input("Enter number of people: "))
tax_percentage = float(input("Enter tax percentage: "))
tip_percentage = float(input("Enter tip percentage: "))

# Step 2: Calculate tax amount
tax_amount = (tax_percentage / 100) * total_bill

# Step 3: Add tax to total bill
bill_after_tax = total_bill + tax_amount

# Step 4: Calculate tip amount on bill after tax
tip_amount = (tip_percentage / 100) * bill_after_tax

# Step 5: Calculate final total bill
final_total = bill_after_tax + tip_amount

# Step 6: Split equally among people
amount_per_person = final_total / number_of_people

# Step 7: Display results
print("\n----- BILL BREAKDOWN -----")
print("Subtotal        :", total_bill)
print("Tax Amount      :", tax_amount)
print("After Tax       :", bill_after_tax)
print("Tip Amount      :", tip_amount)
print("Final Total     :", final_total)
print("Per Person      :", amount_per_person)
```
