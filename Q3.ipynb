```python
# QUESTION NO  3:

# SSTRING MANIPULATION


# Step 1: Ask user to enter a sentence
sentence = input("Enter a sentence: ")

# Step 2: Display original sentence
print("\n----- SENTENCE ANALYSIS -----")
print("1. Original sentence :", sentence)

# Step 3: Total characters (including spaces)
print("2. Total characters (with spaces) :", len(sentence))

# Step 4: Total characters (excluding spaces)
# replace(" ", "") removes spaces from sentence
no_spaces = sentence.replace(" ", "")
print("3. Total characters (without spaces) :", len(no_spaces))

# Step 5: Total words
# split() divides sentence into list of words
words = sentence.split()
print("4. Total words :", len(words))

# Step 6: Convert to uppercase
print("5. UPPERCASE :", sentence.upper())

# Step 7: Convert to lowercase
print("6. lowercase :", sentence.lower())

# Step 8: Convert to Title Case
print("7. Title Case :", sentence.title())

# Step 9: First word
if len(words) > 0:
    print("8. First word :", words[0])
else:
    print("8. First word : No words found")

# Step 10: Last word
if len(words) > 0:
    print("9. Last word :", words[-1])
else:
    print("9. Last word : No words found")

# Step 11: Reversed sentence
# [::-1] reverses the string
print("10. Reversed sentence :", sentence[::-1])
```
