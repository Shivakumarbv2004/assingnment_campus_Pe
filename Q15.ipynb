```python
# QUESTION NO 15:


# PRIME NUMBER CHECKER PROGRAM


# Step 1: Take input
number = int(input("Enter a number: "))

# Step 2: Check prime condition
if number <= 1:
    print("Not a Prime Number")
else:
    is_prime = True

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime Number")
    else:
        print("Not a Prime Number")

```
