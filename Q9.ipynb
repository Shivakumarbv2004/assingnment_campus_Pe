```python
#  QUESTION NO 9:


# TICKET PRICING SYSTEM


# Step 1: Take inputs
age = int(input("Enter age: "))
day = input("Enter day of week: ").lower()
tickets = int(input("Enter number of tickets: "))

# Step 2: Determine base price
if age < 3:
    base_price = 0
elif age <= 12:
    base_price = 150
elif age <= 59:
    base_price = 300
else:
    base_price = 200

# Step 3: Calculate total before discount
total_price = base_price * tickets

# Step 4: Apply weekend discount
if day in ["friday", "saturday", "sunday"]:
    discount = 0.20 * total_price
else:
    discount = 0

final_amount = total_price - discount

# Step 5: Display results
print("\n----- TICKET DETAILS -----")
print("Base Price       :", total_price)
print("Discount         :", discount)
print("Final Amount     :", final_amount)
```
